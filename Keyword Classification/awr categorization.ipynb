```python
import re
a = 'For Rent - General'
re.split(' - ', a)
```

```python
def split_str(url):
    b = re.split('/|-', url)
    my_dict = {'ban': 'For Sale', 'thue': 'For Rent'}
    for i in my_dict:
        if i in b:
            return my_dict.get(i)
    return 'Others'
```

```python
keyword['Parent_Cate'] = keyword['url'].apply(split_str)
```

```python
keyword.head(10)
```

```python
city = {'HCM': 'SG', 'Hanoi':'HN', 'HN':'HN'}
def province(group):
        for a in city:
            if a in re.split(' - |;', group):
                return city.get(a)
        return 'Others'
```

```python
keyword['province'] = keyword['keyword_group'].apply(province)       
```
